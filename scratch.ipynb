```python
teams = pd.read_csv('data/teams.csv')

cols = ['id', 'school', 'mascot', 'classification', 'conference', 'home_venue_id',
        'home_venue', 'venue_capacity', 'city', 'state', 'zip', 'elevation']

teams = teams[teams['classification'] == 'fbs']

teams = teams.dropna(subset=['venue_capacity', 'elevation'])

teams = teams[cols]

teams['venue_capacity_rank'] = teams['venue_capacity'].rank(ascending=False, method='min').astype(int)

teams['elevation_rank'] = teams['elevation'].rank(ascending=False, method='min').astype(int)

teams.to_csv('data/fbs_teams_ranked.csv', index=False)
```

```python
teams = pd.read_csv('data/fbs_teams_with_categories.csv')

teams['smallest_venue_rank'] = teams['venue_capacity'].rank(ascending=True, method='first').astype(int)

teams['largest_venue_rank'] = teams['venue_capacity'].rank(ascending=False, method='first').astype(int)

teams['elevation_rank'] = teams['elevation'].rank(ascending=False, method='first').astype(int)

teams['wins_rank'] = teams['wins'].str.replace(",","").astype(float).rank(ascending=False, method='first').astype(int)

teams['losses_rank'] = teams['losses'].rank(ascending=False, method='first').astype(int)

teams['enrollment_rank'] = teams['enrollment'].str.replace(",","").astype(float).rank(ascending=False, method='first').astype(int)

teams['oldest_rank'] = teams['first_year'].rank(ascending=True, method='first').astype(int)

teams['top_receiver_yards_rank'] = teams['top_receiver_yards'].rank(ascending=False, method='first').astype(int)

teams
```
